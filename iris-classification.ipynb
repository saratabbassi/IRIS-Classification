```python
from sklearn import datasets
iris = datasets.load_iris()
Data=iris.data
Label=iris.target


```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(Data, Label, test_size=0.33, random_state=0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from sklearn.svm import SVC
from sklearn.multiclass import OneVsOneClassifier, OneVsRestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier


# SVM one against one
svm_ovo = OneVsOneClassifier(SVC(kernel='rbf'))
svm_ovo.fit(X_train, Y_train)
svm_ovo_pred = svm_ovo.predict(X_test)
svm_ovo_acc = accuracy_score(Y_test, svm_ovo_pred)
svm_ovo_f1 = f1_score(Y_test, svm_ovo_pred,average='weighted')
svm_ovo_cm = confusion_matrix(Y_test, svm_ovo_pred)

# SVM one against all
svm_ovr = OneVsRestClassifier(SVC(kernel='rbf'))
svm_ovr.fit(X_train, Y_train)
svm_ovr_pred = svm_ovr.predict(X_test)
svm_ovr_acc = accuracy_score(Y_test, svm_ovr_pred)
svm_ovr_f1 = f1_score(Y_test, svm_ovr_pred, average='weighted')
svm_ovr_cm = confusion_matrix(Y_test, svm_ovr_pred)

# Decision tree
dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)
dt_pred = dt.predict(X_test)
dt_acc = accuracy_score(Y_test, dt_pred)
dt_f1 = f1_score(Y_test, dt_pred, average='weighted')
dt_cm = confusion_matrix(Y_test, dt_pred)

# Random forest
rf = RandomForestClassifier()
rf.fit(X_train, Y_train)
rf_pred = rf.predict(X_test)
rf_acc = accuracy_score(Y_test, rf_pred)
rf_f1 = f1_score(Y_test, rf_pred, average='weighted')
rf_cm = confusion_matrix(Y_test, rf_pred)

# XGBoost
xgb = XGBClassifier()
xgb.fit(X_train, Y_train)
xgb_pred = xgb.predict(X_test)
xgb_acc = accuracy_score(Y_test, xgb_pred)
xgb_f1 = f1_score(Y_test, xgb_pred, average='weighted')
xgb_cm = confusion_matrix(Y_test, xgb_pred)

# Print the results
print("SVM one against one accuracy:", svm_ovo_acc)
print("SVM one against one F1-score:", svm_ovo_f1)
print("SVM one against one confusion matrix:\n", svm_ovo_cm)
print("SVM one against all accuracy:", svm_ovr_acc)
print("SVM one against all F1-score:", svm_ovr_f1)
print("SVM one against all confusion matrix:\n", svm_ovr_cm)
print("Decision tree accuracy:", dt_acc)
print("Decision tree F1-score:", dt_f1)
print("Decision tree confusion matrix:\n", dt_cm)
print("Random forest accuracy:", rf_acc)
print("Random forest F1-score:", rf_f1)
print("Random forest confusion matrix:\n", rf_cm)
print("XGBoost accuracy:", xgb_acc)
print("XGBoost F1-score:",xgb_f1)
print("XGBoost confusion matrix:\n:",xgb_cm)
```
